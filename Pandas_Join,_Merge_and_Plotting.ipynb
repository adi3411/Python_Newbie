### 1. Write a Pandas program to append a list of dictioneries or series to an existing DataFrame and display the combined data.

```python
import pandas as pd

# Create the initial DataFrame
student_detail = pd.DataFrame({
    'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
    'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'],
    'marks': [200, 210, 190, 222, 199]
})

# New student dictionary
new_student = {
    'student_id': 'S6',
    'name': 'Scarlette Fisher',
    'marks': 205
}

# Convert the dictionary into a DataFrame
new_student_df = pd.DataFrame([new_student])

# Concatenate the original DataFrame with the new one
student_detail = pd.concat([student_detail, new_student_df], ignore_index=True)

# Display the combined data
student_detail
```

### 2. Write a Pandas program to join the two given dataframes a long rows and merge with another dataframe along the common column id.

```python
import pandas as pd

# Creating first student data as a dataframe
data1 = {
    'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
    'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'],
    'marks': [200, 210, 190, 222, 199]
}
student_data1 = pd.DataFrame(data1)

# Creating second student data as a dataframe
data2 = {
    'student_id': ['S6', 'S7', 'S8', 'S9', 'S10'],
    'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'],
    'marks': [201, 200, 198, 219, 201]
}
student_data2 = pd.DataFrame(data2)

# Joining both of the student dataframes along the rows
combined_student_data = pd.concat([student_data1, student_data2], ignore_index=True)


# Create the exam data dataframe
exam_data = {
    'student_id': ['S1', 'S2', 'S3', 'S4', 'S5', 'S7', 'S8', 'S9', 'S10', 'S11', 'S12', 'S13'],
    'exam_id': [23, 45, 12, 67, 21, 55, 33, 14, 56, 83, 88, 12]
}
exam_data = pd.DataFrame(exam_data)

# Merge the combined student data with the exam data on 'student_id'
merged_data = pd.merge(combined_student_data, exam_data, on='student_id', how='left')

# Display the final merged dataframe
print('The final merged dataframe is as below: ')
print('')
merged_data
```

###3. Write a Pandas program to join the two dataframes using the common column of both dataframes

```python
import pandas as pd

#first dataframe
data1 = {
    'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
    'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'],
    'marks': [200, 210, 190, 222, 199]
}

student_data1 = pd.DataFrame(data1)

#second dataframe
data2= {
    'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
    'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'],
    'marks': [201, 200, 198, 219, 201]
}

student_data2 = pd.DataFrame(data2)

# Merge the two dataframes on the common column 'student_id'
merged_data = pd.merge(student_data1, student_data2, on='student_id', how='inner', suffixes=('_data1', '_data2'))

# Display the merged dataframe
print('The final merged dataframe on the common column is as below: ')
print('')
merged_data
```

###4. Write a Pandas program to join the two dataframes with matching records from both sides where available.

```python
import pandas as pd

# Creating the first DataFrame
student_data1 = pd.DataFrame({
    'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
    'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'],
    'marks': [200, 210, 190, 222, 199]
})

# Creating the second DataFrame
student_data2 = pd.DataFrame({
    'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
    'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'],
    'marks': [201, 200, 198, 219, 201]
})

# Merging the two DataFrames on 'student_id' with an inner join
merged_data = pd.merge(student_data1, student_data2, on='student_id', how='inner', suffixes=('_data1', '_data2'))

# Display the result
merged_data
```

###5. Write a Pandas program to join (left join) the two dataframes using keys from left dataframe only.

```python
import pandas as pd

# Creating the first dataframe (data1)
data1 = {
    'key1': ['K0', 'K0', 'K1', 'K2'],
    'key2': ['K0', 'K1', 'K1', 'K1'],
    'P': ['P0', 'P1', 'P2', 'P3'],
    'Q': ['Q0', 'Q1', 'Q2', 'Q3']
}
df1 = pd.DataFrame(data1)

# Creating the second dataframe (data2)
data2 = {
    'key1': ['K0', 'K1', 'K2', 'K2'],
    'key2': ['K0', 'K1', 'K0', 'K1'],
    'R': ['R0', 'R1', 'R2', 'R3'],
    'S': ['S0', 'S1', 'S2', 'S3']
}
df2 = pd.DataFrame(data2)

# Performing a left join using keys from the left dataframe
result = pd.merge(df1, df2, on=['key1', 'key2'], how='left')

# Displaying the result
result
```

###6. Write a Pandas program to create a line plot of the historical stock prices of Alphabet Inc. between two specific dates.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/drive/MyDrive/DB & Stats/alphabet_stock_data.csv')
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

plt.plot(data.loc['2020-09-01':'2020-09-30'].index, data['Close'].loc['2020-09-01':'2020-09-30'])
plt.title('Historical Stock Prices of Alphabet Inc.')
plt.xlabel('Date')
plt.ylabel('Stock Price (USD)')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

###7. Write a Pandas program to create a line plot of the opening, closing stock prices of Alphabet Inc. between two specific dates.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/DB & Stats/alphabet_stock_data.csv')

# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Set the 'Date' column as the index
data.set_index('Date', inplace=True)

# Filter data between two specific dates
filtered_data = data.loc['2020-09-01':'2020-09-30']

# Create a line plot for the 'Open' and 'Close' prices
plt.plot(filtered_data.index, filtered_data['Open'], label='Opening Price', color='red')
plt.plot(filtered_data.index, filtered_data['Close'], label='Closing Price', color='blue')

# Add title and labels
plt.title('Opening and Closing Stock Prices of Alphabet Inc.')
plt.xlabel('Date')
plt.ylabel('Stock Price (USD)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend()  # Show legend to differentiate between opening and closing prices
plt.grid()  # Add grid for better visualization
plt.show()  # Display the plot
```

###8. Write a Pandas program to create a bar plot of the trading volume of Alphabet Inc. stock between two specific dates.

```python
# Create a bar plot for the trading volume
plt.bar(filtered_data.index, filtered_data['Volume'], color='orange')

# Add title and labels
plt.title('Trading Volume of Alphabet Inc. Stock')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y')  # Add grid for better visualization on y-axis
plt.show()  # Display the plot
```

###9. Write a Pandas program to create a bar plot of opening, closing stock prices of Alphabet Inc. between two specific dates.

```python
# Create a bar plot for the 'Open' and 'Close' prices
plt.figure(figsize=(11, 6))

bar_width = 0.35
x = range(len(filtered_data))

plt.bar(x, filtered_data['Open'], width=bar_width, label='Opening Price', color='blue', align='center')
plt.bar([p + bar_width for p in x], filtered_data['Close'], width=bar_width, label='Closing Price', color='red', align='center')

# Add title and labels
plt.title('Opening and Closing Stock Prices of Alphabet Inc.')
plt.xlabel('Date')
plt.ylabel('Stock Price (USD)')
plt.xticks([p + bar_width / 2 for p in x], filtered_data.index.date, rotation=45)  # Set x-ticks to dates
plt.legend()  # Show legend to differentiate between opening and closing prices
plt.grid(axis='y')  # Add grid for better visualization on y-axis
plt.show()  # Display the plot
```

### 10. Write a Pandas program to create a stacked bar plot of opening, closing stock prices of Alphabet Inc. between two specific dates.

```python
# Create a stacked bar plot for the 'Open' and 'Close' prices
plt.figure(figsize=(14, 6))  # Set the figure size to be wider

plt.bar(filtered_data.index, filtered_data['Open'], label='Opening Price', color='orange')
plt.bar(filtered_data.index, filtered_data['Close'], bottom=filtered_data['Open'], label='Closing Price', color='blue')

# Add title and labels
plt.title('Stacked Bar Plot of Opening and Closing Stock Prices of Alphabet Inc.')
plt.xlabel('Date')
plt.ylabel('Stock Price (USD)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend()  # Show legend to differentiate between opening and closing prices
plt.grid(axis='y')  # Add grid for better visualization on y-axis
plt.show()  # Display the plot
```
