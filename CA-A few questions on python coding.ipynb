# Question 1: Real World Tax Calculation System

```python
'''Creates a Income tax calcuator taking in inputs required for calculating the tax liability
   and basis the inputs provided, calculates the final tax liability.
   Defines functions to calculate USC, PRSI, Income Tax and Tax Credits
'''

#Defining Universal Social Charge(USC)Slabs:
''' Calculates the Universal Social charge slabs, based on gross income'''
def USC(gross_income):
  if gross_income <= 12012:
    USC = gross_income * 0.005
  elif gross_income <= 27382:   #(12012 + 15370)
    USC = ((12012 * 0.005) + ((gross_income - 12012) * 0.02))
  elif gross_income <= 70044:   #(12012 + 15370 + 42662)
    USC = (12012 * 0.005) + (15370 * 0.02) + ((gross_income - 27382) * 0.03)
  else:                         #if remaining balance is > 42662
    USC = (12012 * 0.005) + (15370 * 0.02) + (42662 * 0.03) + ((gross_income - 70044) * 0.08)
  return USC

#Defining Income tax slabs
''' Calculates the Income tax slabs, based on gross income'''
def income_tax(gross_income):
  balance = gross_income - 44000
  if gross_income <= 44000:
    return gross_income * 0.2
  else:
    return (44000 * 0.2) + (balance * 0.4)

#Defining & calculating Tax credits
''' Calculates the Tax credits, based on a few factors such as single or married,
    if employee has credit, rent credit and total number of dependents
'''
def tax_credits(Single, emp_credit, rent_credit, no_dep):
  if Single == 'yes':
    credits = 2000
  else:
    credits = 4000
  if emp_credit == 'yes':
    credits += 2000
  else:
    credits += 0
  if rent_credit == 'yes':
    credits += 1000
  else:
    credits += 0
  if no_dep == '0':
    credits += 0
  elif no_dep == '1':
    credits += 305
  elif no_dep == '2':
    credits += 305 * 2
  else:
    credits += 0
  return credits


#Taking User Inputs
''' Taking the user inputs and convert to lower to keep it consistent throughout the code'''
#1. Employee ID
Emp_ID = input("Plese Enter your Empployee ID (Alphanumeric): ")

#2. Employee Name
Emp_Name = input("Please enter your Name: ")
if all(char.isalpha() or char.isspace() for char in Emp_Name):
    pass
    #print("Employee Name: ", Emp_Name)
else:
  print("Error: Employee Name should only contain letters and spaces.")

#3. Gross Income
GI = input("Enter your yearly Gross Income: ")
if all(char.isnumeric() for char in GI):
  gross_income = abs(float(GI))
else:
  print("Error: Gross Income should only contain NUmericals.")

#4. Single - YEs/NO
Single = input("Are you a Single? (Yes/No): ").lower()
if Single == "yes" or Single == "no":
  pass
else:
  print("Error: Please enter Yes or No")

#5. Married/Civil Partner - Yes/No
Married = input("Are you a Married/Civil Partner? (Yes/No): ").lower()
if Single == 'yes':
  Married = 'No'
elif Single == 'no':
  Married = 'Yes'
else:
  print("Error: Please enter Yes or No")

print(Married)

#6. Employee Credit - Yes/No
emp_credit = input("Do you have any Employee Credit? (Yes/No): ").lower()
if emp_credit == "yes" or emp_credit == "no":
  pass
else:
  print("Error: Please enter Yes or No")

#7. Rent credit - Yes.-no
rent_credit = input("Do you have any Rent Credit? (Yes/No): ").lower()
if rent_credit == "yes" or rent_credit == "no":
  pass
else:
  print("Error: Please enter Yes or No")

#8. No. of Dependents - 0 or 1 or 2
no_dep = input("Enter the number of Dependents (either 0 or 1 or 2 only): ")
if no_dep == "0" or no_dep == "1" or no_dep == "2":
  pass
else:
  print("Error: It must be 0, 1 or 2 only")


#User Inputs
'''For ease of access, showcase all the inputs at a glance '''
print("-------------------------------------------")
print("Inputs at a glance...")
print("Employee ID is: ", Emp_ID)
print("Employee Name is: ", Emp_Name)
print("His Gross Income is: ",gross_income)
print("Employee Single or Not: " ,Single)
print("Employee is Married or Not: ",Married)
print("Employee has ccredit: ", emp_credit)
print("Employee has Rent ccredit: ",rent_credit)
print("Total number of dependents: ",no_dep)
print("-------------------------------------------")

#Caclulating final tax liability
USC = USC(gross_income)
PRSI = gross_income * 0.041
Income_Tax = income_tax(gross_income)
Tax_Credits = tax_credits(Single, emp_credit, rent_credit, no_dep)
Final_Tax = (USC + PRSI + Income_Tax) - Tax_Credits
Net_Income = gross_income - Final_Tax

#Outputs Values
print("________________________________________________________________")
print("")
print(" Statement of Liablity for ", Emp_Name)
print("________________________________________________________________")
print("")
print("Gross Income is: ", "€",gross_income)
print("USC Deduction is: ", "€",USC)
print("PRSI Deduction is: ", "€",PRSI)
print("Income Tax is: ", "€",Income_Tax)
print("________________________________________________________________")
print("")
print("Total Tax credits applied: ", "€",Tax_Credits)
print("Final Tax Liability: ", "€",Final_Tax)
print("________________________________________________________________")
print("")
print("Net Income: ", "€",Net_Income)
print("________________________________________________________________")
print("")

```

## Question 2: Inventory Management System

```python
#Inventory Management System
''' This code makes and invesntory Management system with option such as Add, Delete, Update and Look-up functions
'''
#As we donot have any databse, we have to initialize an empty dictionayr first
inv_mgmt_system = {}

#defining the function to ADD a new product to the inventory
def add_product():
  '''This function ADDS a new product to the inventory - after checking if anything with the same name already exists.
     To keep it consistent get the input in upper case
  '''
  product_ID = input('Enter the ID of the product: ')
  product_name = input('Enter the name of the product, (Can be alpha-numeric): ')
  product_name = product_name.upper()   #this is so that we have no case sensitive error and no duplicate to be added.
  if product_name in inv_mgmt_system:
    print('Product with that name already exists in the inventory')
  else:
    print('')
    print('--------------------------------------------------------')
    while True:
      try:
        quantity = int(input('Enter the quantity of the product: '))
        if quantity < 0:
          print('Error! Enter a positive number value')
        else:
         break
      except ValueError:
        print('Enter a numerical value')
    print('')
    while True:
      try:
        price = float(input('Enter the price of the product: '))
        if price < 0:
          print('Error! Enter a positive number value')
        else:
         break
      except ValueError:
        print('Enter a numerical value')
    print('')
    inv_mgmt_system[product_name] = {'quantity': quantity, 'price': price}
    print('Product ', product_name, 'has been added to the inventory sucessfully')
    print('--------------------------------------------------------')

#defining the function to DELETE a product from the inventory
'''This function DELETES the exisitng product to the inventory - after checking if anything with the same name already exists.
     To keep it consistent get the input in upper case
     Added a warning to confirm the deletion
  '''
def delete_product():
  product_ID = input('Enter the ID of the product: ')
  product_name = input("Enter the name of the Product to be deleted from the inventory: ")
  product_name = product_name.upper()
  print('')
  print('--------------------------------------------------------')
  print('Warning!')
  print('Once deleted the entity can not be recovered.')
  print('')
  choice = input('Are you sure you want to delete this product? (Yes/No): ').lower()
  if choice == 'yes':
    if product_name in inv_mgmt_system:
      del inv_mgmt_system[product_name]
      print('Product ', product_name, 'has been deleted from the inventory sucessfully')
    else:
      print('')
      print('Product with that name does not exist in the inventory')
  else:
    print('Deletion Cancelled')
  print('--------------------------------------------------------')

#defining the function to UPDATE the existing product to the inventory
'''This function UPDATES the exisitng product to the inventory - after checking if anything with the same name already exists.
     To keep it consistent get the input in upper case
  '''
def update_product():
  product_ID = input('Enter the ID of the product: ')
  product_name = input('Enter the name of the product to be updated: ')
  product_name = product_name.upper()
  if product_name in inv_mgmt_system:
    print('')
    print('--------------------------------------------------------')
    while True:
      try:
        quantity = int(input('Enter the new quantity of the product: '))
        if quantity < 0:
          print('Error! Enter a positive number value')
        else:
         break
      except ValueError:
        print('Enter a numerical value')
    print('')
    while True:
      try:
        price = float(input('Enter the new price of the product: '))
        if price < 0:
          print('Error! Enter a positive number value')
        else:
         break
      except ValueError:
        print('Enter a numerical value')
    print('')
    inv_mgmt_system[product_name] = {'quantity': quantity, 'price': price}
    print('Product ', product_name, 'has been updated sucessfully')
    print('--------------------------------------------------------')
  else:
    print("Product not found.")

#defining the function to LOOK-UP an existing product to the inventory
'''This function LOOKs-UP the exisitng product in the inventory - after checking if anything with the same name already exists.
     To keep it consistent get the input in upper case
     Added an option to add the new product if not found
      and also added an option to confirm the Cancellation
  '''
def lookup_product():
  product_ID = input('Enter the ID of the product: ')
  product_name = input("Enter the product name to look up from the inventory: ")
  product_name = product_name.upper()
  if product_name in inv_mgmt_system:
    product = inv_mgmt_system[product_name]
    print('')
    print('--------------------------------------------------------')
    print('Product ID: ', product_ID)
    print('Product Name: ', product_name)
    print('Quantity: ', product['quantity'])
    print('Price: ', product['price'])
  else:
    print('')
    print('There is no product with that name!')
    add_new = input('Would you like to add this product? (Yes/No): ')
    add_new = add_new.lower()
    if add_new == 'yes':
      add_product()
    else:
      print('Addition of new product cancelled!')
    print('--------------------------------------------------------')

#Inputs and choice
while True:
  print('')
  print('_______ Inventory Management System________')
  print('')
  print('1. "Add" a product to the invesntory')
  print('2. "Delete" a product from the inventory')
  print('3. "Update" the existing product in the invesntory')
  print('4. "Lookup"/find the product from the inventory')
  print('5. Exit')
  print('___________________________________________')
  print('')
  selection = input('Enter your choice (1-5): ')

  if selection == '1':
    add_product()
  elif selection == '2':
    delete_product()
  elif selection == '3':
    update_product()
  elif selection == '4':
    lookup_product()
  elif selection == '5':
    break
  else:
    print('Invalid choice. Please enter a number between 1 and 5.')

  #references
  #W3Schools, 2023. Python Try Except. Available at: https://www.w3schools.com/python/python_try_except.asp
```

##Question 3.  Body Mass Index (BMI) Calculator

```python
""" BMI calcs:
Calculating the BMI from the
"""

#Converting Pounds to Kilograms(Kg)
def to_kg(weight_in_pounds):
  return weight_in_pounds * 0.4535

#Converting Inches to Meters
def to_meters(height_in_inches):
  return height_in_inches * 0.0254

#defining the calculaton for BMI
def BMI_main(weight_in_kg, height_in_meters):
  bmi = weight_in_kg / (height_in_meters ** 2)
  return round(bmi, 2) #rounding to 2 decimal places.

#defining the BMI categories here
def BMI_category(bmi):
  if bmi < 18.5:
    return "You are Underweight"
  elif 81.5 <= bmi < 25:
    return "You are Healthy"
  elif 25 <= bmi < 30:
    return "You are Overweight"
  else:
    return "You are Obese"


#main inouts - in decimals
weight_in_pounds = float(input("Enter weight in Pounds: "))
height_in_inches = float(input("Enter height in Inches: "))

#Calling converted weight to KG from pounds
weight_in_kg = to_kg(weight_in_pounds)

#calling Converted height to Meters from inches
height_in_inches = to_meters(height_in_inches)

#caling BMI
BMI = BMI_main(weight_in_kg, height_in_inches)

#outputs
print('')
print('_______BMI Calculation_______')
print('____________________________________')
print('')
print('Your weight in KG is: ', weight_in_kg)
print('Your height in Meters is: ', height_in_inches)
print('')
print('Your BMI is: ', BMI, 'kg/m2')
print('You are: ', BMI_category(BMI))
print('____________________________________')

```

## Question 4. String Reversal without reversing its individual words

```python
""" String Reversing to verify if the enterd stringis empty or not
if not empty, then reverse the input string, by spliting and reversing.
in the meantime, once split, count the the length of each word
"""

#defining a funcation to reverse the words in a string
def rev_string_count(string):
  if not string.strip():   #verifiy if the entered string is empty or not
    return ""

  words = string.split()    #Split the string into words - remove exxtra spaces

  reversed_words = words[::-1]     #reverse the split words

  word_lenght = [(word, len(word)) for word in reversed_words]  #creating a list to to hold the word and their lenght as tuple - else it would pose an error

  reversed_string = ' '.join(reversed_words)

  return reversed_string, word_lenght   #return the reversed string and their lengths


#getting the input in a string
string = input('Enter a sentence. Use spaces to separate words & enclose in quotations: ')

#output
print('')
reversed_string, word_lenght = rev_string_count(string)
print('----------------------------------------------------------------------------------------------')
print('_______Output - String Reversal________')
print('')
print('Length of each word: ', word_lenght)
print('')
print('Reversed String with same Punctuation and characters is: ', reversed_string)
print('----------------------------------------------------------------------------------------------')

#References
#W3Schools. (n.d.). Python String split() Method. Available at: https://www.w3schools.com/python/ref_string_split.asp
#Python Software Foundation, 2023. Data Structures. Available at: https://docs.python.org/3/tutorial/datastructures.html
#W3Schools, 2023. Python Tuples. Available at: https://www.w3schools.com/python/python_tuples.asp
#Programiz, 2023. Python List Comprehension. Available at: https://www.programiz.com/python-programming/list-comprehension
```

```python
""" Takingg user input od desired stock price
by callingin yFinance and datetime library.
"""

import yfinance as yf    #yfinance API is used to access stock data and tickers
from datetime import datetime, timedelta   #datetime is used to handle date calculations

#asking for user inputs

print('')
while True:
  try:
    buy = float(input("At what price do you wish to buy this stock? Enter Numerical data only: "))
    break
  except ValueError:
    print('Enter a Numerical ')

#SHowing the stock prince for 7 days -
#select the enddate as of now and calcualte 7 days backwords - as stock prices are past records
end_date = datetime.now()
start_date = end_date - timedelta(days = 7)
#fetching the details of stock Tesla via yfinance API
stock = yf.Ticker("TSLA")
stock_data = stock.history(start = start_date, end = end_date)

#Showcasing the Tesla stock prices
print('_______________________________________________')
print('')
print('The last 7 days Closing prices for Tesla are: ')
print('_______________________________________________')
print('')
for date, row in stock_data.iterrows():
  print(f"Date: {date.date()}  -  Closing Price: {row['Close']:.4f}")
print('_______________________________________________')

#Checking if the buy price  is above the current pricce
today_price = stock_data['Close'].iloc[-1]
if buy > today_price:
  print('')
  print('Conclusion! ')
  print('You can buy this stock at your desired price, as the last closing price is teh table above')
else:
  print('')
  print('Conclusion! ')
  print('You will have to wait for a better price, as current price is higher than your price (refer to the prices above)')

#References
#https://docs.python.org/3/tutorial/errors.html
#https://www.w3schools.com/python/python_user_input.asp
#Juunge, K. (2021). yfinance: 10 Ways to Get Stock Data with Python. Available at: https://medium.com/@kasperjuunge/yfinance-10-ways-to-get-stock-data-with-python-6677f49e8282
#Python Programming. (n.d.). Getting Stock Prices - Python Programming for Finance. Available at: https://pythonprogramming.net/getting-stock-prices-python-programming-for-finance/
```
